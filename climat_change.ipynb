```python
import pandas as pd
df = pd.read_csv("Environment_Temperature_change.csv", delimiter=',',encoding = "Windows-1251")

cols = [0,2,4,6]
df.drop (df.columns [cols], axis= 1 , inplace= True )

df = df[(df.Months == 'Meteorological year') & (df.Element != 'Standard Deviation')]
df
```

```python
a = df.groupby(['Y1965']).agg({'Area':['max']})
a = a.iloc [[0,-1]]
a = a.sort_values(by=['Y1965'], ascending=[False])
a
```

```python
df.to_csv (r' Environment_Temperature_change_for_lab3.csv', index= False )
```

SELECT Area, Months, max(Value) FROM tempsheet GROUP BY Area, Months, Year

```python
import pandas as pd
df1 = pd.read_csv("gg.csv", delimiter=',',encoding = "Windows-1251")
df1 = df1.dropna ()

cols = [0,1,2,4,5,6,8,10,12,13]
df1.drop (df1.columns [cols], axis= 1 , inplace= True )

df1 = df1[(df1.Months != 'Dec?Jan?Feb') & 
          (df1.Months != 'Mar?Apr?May') & 
          (df1.Months != 'Jun?Jul?Aug') & 
          (df1.Months != 'Sep?Oct?Nov') & 
          (df1.Months != 'Meteorological year')]

a12 = df1.groupby(['Area', 'Months', 'Year']).agg({'Value':'max'})
a12 = a12.sort_values(by=['Area'], ascending=[True])
a12
```

```python
df1.to_csv (r' gg1.csv', index= False )
```
