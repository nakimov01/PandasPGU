-Помещение файла в датафрейм

```python
import pandas as pd
df = pd.read_csv("taxi_peru.csv", delimiter=';', parse_dates=['start_at', 'end_at'], encoding = "Windows-1251")
df
```

```python
df.dtypes
```

Добавление значений в датафрейм new_df из датафрейма df. Переименование значений journey_id, source, start_type

```python
new_df = df[['journey_id', 'user_id', 'driver_id','rider_score','driver_score', 'source', 'start_type']].rename(columns={'journey_id':'order_id', 'source':'user_platform', 'start_type':'order_type'})
new_df
```

Добавление колонки с общим временем поездки в датафрейм new_df

```python
from datetime import datetime, timedelta

new_df['travel_time_in_hours'] = (df['end_at']-df['start_at'])

new_df['travel_time_in_hours'] = (new_df['travel_time_in_hours'] / pd.Timedelta(hours=1)).round(2)

new_df['travel_time_in_hours'] = new_df.loc[(new_df['travel_time_in_hours'] > 0.1), 'travel_time_in_hours']

new_df
```

Подготовка координат в правильный формат и дальнейшая конвертация из str в float для нахождения растояния поездки

```python
df['start_lat'] = df['start_lat'].astype(str).str.replace(',', '.')
df['end_lat'] = df['end_lat'].astype(str).str.replace(',', '.')
df['start_lon'] = df['start_lon'].astype(str).str.replace(',', '.')
df['end_lon'] = df['end_lon'].astype(str).str.replace(',', '.')
df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]]= df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]].astype(float)
df
```

Вычисление расстояния по координатам из датафрейма df и создание колонки travel_distance с дальнейшим перемещением в new_df

```python
import numpy as np
new_df['travel_distance'] = df.apply(lambda df: ((np.sqrt((df['end_lat']-df['start_lat'])**2 + (df['end_lon']-df['start_lon'])**2))*100), axis=1).round(3)
new_df['travel_distance'] = new_df.loc[(new_df['travel_distance'] > 0.1) & (new_df['travel_distance'] < 100), 'travel_distance']
new_df
```

Определение лучшего водителя по 3 колонкам(количеству заказов, средней оценке, общему количеству пройденных километров)

```python
Rename_columns = new_df[['driver_id', 'order_id', 'driver_score', 'travel_distance']]
GroupBy_drivers = Rename_columns.groupby(['driver_id']).agg({'order_id':'count', 'driver_score':'mean', "travel_distance":'sum'})
best_driver = GroupBy_drivers.sort_values(by=['order_id', 'driver_score', 'travel_distance'], ascending=[False, False, False]).round(2)

max_sum_orders_value = best_driver['order_id'].max()
max_mean_driver_score_value = 5.0
max_travel_distance_value = best_driver['travel_distance'].max()

best_driver['result'] = best_driver.apply(lambda best_driver: ((best_driver['order_id']/max_sum_orders_value)
                                                               + (best_driver['driver_score']/max_mean_driver_score_value)
                                                               + (best_driver['travel_distance']/max_travel_distance_value)), axis=1).round(5)

best_driver.sort_values(by=['result'], ascending=[False]).head(10).round(3)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
best_driver = best_driver.sort_values(by=['result'], ascending=[False]).head(11)
sns.barplot(x = best_driver.result, y = best_driver.index , data = best_driver)
```

Определение лучшего пользователя по 3 колонкам(количеству поездок, средняя оценка водителя, общему количеству километров в поездках)

```python
Rename_columns = new_df[['user_id', 'order_id', 'rider_score', 'travel_distance']]
GroupBy_drivers = Rename_columns.groupby(['user_id']).agg({'order_id':'count', 'rider_score':'mean', "travel_distance":'sum'})
best_rider = GroupBy_drivers.sort_values(by=['order_id', 'rider_score', 'travel_distance'], ascending=[False, False, False]).round(2)

max_sum_orders_value = best_rider['order_id'].max()
max_mean_rider_score_value = 5.0
max_travel_distance_value = best_rider['travel_distance'].max()

best_rider['result'] = best_rider.apply(lambda best_rider: ((best_rider['order_id']/max_sum_orders_value) 
                                                            + (best_rider['rider_score']/max_mean_rider_score_value) 
                                                            + (best_rider['travel_distance']/max_travel_distance_value)), axis=1).round(5)

best_rider.sort_values(by=['result'], ascending=[False]).head(10).round(5)
```

```python
best_rider = best_rider.sort_values(by=['result'], ascending=[False]).head(10)
sns.barplot(x = best_rider.result, y = best_rider.index , data = best_rider)
```
