-Помещение файла в датафрейм
-Изменение в рейтинге со значения NaN на float 0.0
-Удаление лишних колонок со значением 'driver cancel', 'not found', 'rider cancel'

```python
import pandas as pd
df = pd.read_csv("taxi_peru.csv", delimiter=';', parse_dates=['start_at', 'end_at'], encoding = "Windows-1251")
df['rider_score'] = df['rider_score']. fillna (0)
df = df[(df.end_state == 'drop off')]
df
```

```python
df.dtypes
```

Сколько уникальных значений такси и водителей имеется в датафрейме

```python
df.taxi_id.unique().shape
df.driver_id.unique().shape
```

Подсчет количества водителей и такси на которых они ездили

```python
a = df.groupby(['driver_id']).agg({'taxi_id':pd.Series.nunique, 'journey_id':'count'})
a.sort_values(by=['journey_id'], ascending=[False])
```

Добавление значений в датафрейм new_df из датафрейма df. Переименование значений journey_id, source, start_type

```python
new_df = df[['journey_id', 'user_id', 'driver_id','rider_score','driver_score', 'source', 'start_type']].rename(columns={'journey_id':'order_id', 'source':'user_platform', 'start_type':'order_type'})
new_df
```

Добавление колонки с общим временем поездки в датафрейм new_df

```python
from datetime import datetime, timedelta

new_df['travel_time_in_hours'] = (df['end_at']-df['start_at'])

new_df['travel_time_in_hours'] = (new_df['travel_time_in_hours'] / pd.Timedelta(hours=1)).round(2)

new_df
```

Подготовка координат в правильный формат и дальнейшая конвертация из str в float для нахождения растояния поездки

```python
df['start_lat'] = df['start_lat'].astype(str).str.replace(',', '.')
df['end_lat'] = df['end_lat'].astype(str).str.replace(',', '.')
df['start_lon'] = df['start_lon'].astype(str).str.replace(',', '.')
df['end_lon'] = df['end_lon'].astype(str).str.replace(',', '.')
df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]]= df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]].astype(float)
df
```

Вычисление расстояния по координатам из датафрейма df и создание колонки travel_distance с дальнейшим перемещением в new_df

```python
import numpy as np
new_df['travel_distance'] = df.apply(lambda df: ((np.sqrt((df['end_lat']-df['start_lat'])**2 + (df['end_lon']-df['start_lon'])**2))*100), axis=1)
new_df
```

Определение лучшего водителя по 4 колонкам(количеству заказов, средняя оценка, сумма часов работы, общее количество пройденных километров)

```python
Sort = new_df.query("`travel_distance` < 100 and `travel_distance` > 0.25 ")
a = Sort.groupby(['driver_id']).agg({'order_id':'count', 'driver_score':'mean', 'travel_time_in_hours':'sum', "travel_distance":'sum'})
a.sort_values(by=['order_id', 'driver_score', 'travel_time_in_hours','travel_distance'], ascending=[False, False, False, False]).round(2).head(5)
```

Определение лучшего пользователя по 3 колонкам(количеству поездок, средняя оценка водителем, общее количество километров в поездках)

```python
Sort = new_df.query("`travel_distance` < 100 and `travel_distance` > 0.25 ")
a = Sort.groupby(['user_id']).agg({'order_id':'count', 'rider_score':'mean', 'travel_distance':'sum'})
a.sort_values(by=['order_id', 'rider_score', 'travel_distance'], ascending=[False, False, False]).round(2)
```
