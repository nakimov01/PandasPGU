```python
import pandas as pd
df = pd.read_csv("taxi_peru.csv", delimiter=';', parse_dates=['start_at', 'end_at'], encoding = "Windows-1251")
df = df[(df.end_state == 'drop off')]
df
```

Добавление значений в датафрейм new_df из датафрейма df. Переименование значений journey_id, source, start_type

```python
new_df = df[['journey_id', 'user_id', 'driver_id','rider_score','driver_score', 'source', 'start_type']].rename(columns={'journey_id':'order_id', 'source':'user_platform', 'start_type':'order_type'})
new_df
```

Добавление колонки с общим временем поездки в датафрейм new_df

```python
from datetime import datetime, timedelta

new_df['travel_time_in_hours'] = (df['end_at']-df['start_at'])

new_df['travel_time_in_hours'] = (new_df['travel_time_in_hours'] / pd.Timedelta(hours=1)).round(2)

new_df['travel_time_in_hours'] = new_df.loc[(new_df['travel_time_in_hours'] > 0.1), 'travel_time_in_hours']

new_df
```

Подготовка координат в правильный формат и дальнейшая конвертация из str в float для нахождения растояния поездки

```python
df['start_lat'] = df['start_lat'].astype(str).str.replace(',', '.')
df['end_lat'] = df['end_lat'].astype(str).str.replace(',', '.')
df['start_lon'] = df['start_lon'].astype(str).str.replace(',', '.')
df['end_lon'] = df['end_lon'].astype(str).str.replace(',', '.')
df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]]= df[['start_lat', 'end_lat', 'start_lon', 'end_lon', ]].astype(float)
df
```

Вычисление расстояния по координатам из датафрейма df и создание колонки travel_distance с дальнейшим перемещением в new_df

```python
import numpy as np
new_df['travel_distance'] = df.apply(lambda df: ((np.sqrt((df['end_lat']-df['start_lat'])**2 + (df['end_lon']-df['start_lon'])**2))*100), axis=1).round(3)
new_df['travel_distance'] = new_df.loc[(new_df['travel_distance'] > 0.00) & (new_df['travel_distance'] < 100), 'travel_distance']
new_df
```

Определение лучшего водителя по 3 колонкам(количеству заказов, средней оценке, общему количеству пройденных километров)

```python
Rename_columns = new_df[['driver_id', 'order_id', 'driver_score', 'travel_distance']].rename(columns={'order_id':'order_ids_points', 'driver_score':'driver_score_points','travel_distance':'travel_distance_points'})
GroupBy_drivers = Rename_columns.groupby(['driver_id']).agg({'order_ids_points':'count', 'driver_score_points':'mean', "travel_distance_points":'sum'})
best_driver = GroupBy_drivers.sort_values(by=['order_ids_points', 'driver_score_points', 'travel_distance_points'], ascending=[False, False, False]).round(2)

best_driver.loc[(best_driver['order_ids_points'] <= 200), 'order_ids_points'] = best_driver['order_ids_points'] * 1.5
best_driver.loc[(best_driver['order_ids_points'] > 200) & (best_driver['order_ids_points'] <= 400), 'order_ids_points'] = best_driver['order_ids_points'] * 2.0
best_driver.loc[(best_driver['order_ids_points'] > 400) & (best_driver['order_ids_points'] <= 600), 'order_ids_points'] = best_driver['order_ids_points'] * 2.5
best_driver.loc[(best_driver['order_ids_points'] > 600) & (best_driver['order_ids_points'] <= 800), 'order_ids_points'] = best_driver['order_ids_points'] * 3
best_driver.loc[(best_driver['order_ids_points'] > 800) & (best_driver['order_ids_points'] <= 1000), 'order_ids_points'] = best_driver['order_ids_points'] * 3.50
best_driver.loc[(best_driver['order_ids_points'] > 1000) & (best_driver['order_ids_points'] <= 1500), 'order_ids_points'] = best_driver['order_ids_points'] * 4

best_driver.loc[(best_driver['driver_score_points'] <= 4.00), 'driver_score_points'] = best_driver['driver_score_points'] * 10
best_driver.loc[(best_driver['driver_score_points'] > 4.00) & (best_driver['driver_score_points'] <= 4.25), 'driver_score_points'] = best_driver['driver_score_points'] * 15
best_driver.loc[(best_driver['driver_score_points'] > 4.25) & (best_driver['driver_score_points'] <= 4.50), 'driver_score_points'] = best_driver['driver_score_points'] * 20
best_driver.loc[(best_driver['driver_score_points'] > 4.50) & (best_driver['driver_score_points'] <= 4.75), 'driver_score_points'] = best_driver['driver_score_points'] * 30
best_driver.loc[(best_driver['driver_score_points'] > 4.75) & (best_driver['driver_score_points'] <= 4.90), 'driver_score_points'] = best_driver['driver_score_points'] * 40
best_driver.loc[(best_driver['driver_score_points'] > 4.90), 'driver_score_points'] = best_driver['driver_score_points'] * 50

best_driver['total_points'] = (best_driver['order_ids_points']+best_driver['driver_score_points']+best_driver['travel_distance_points'])

best_driver.sort_values(by=['total_points'], ascending=[False])

best_driver.head(11)
```

Определение лучшего пользователя по 3 колонкам(количеству поездок, средняя оценка водителя, общему количеству километров в поездках)

```python
Rename_columns2 = new_df[['user_id', 'order_id', 'rider_score', 'travel_distance']].rename(columns={'order_id':'order_ids_points', 'rider_score':'rider_score_points',
                                                                                                    'travel_distance':'travel_distance_points'})
GroupBy_riders = Rename_columns2.groupby(['user_id']).agg({'order_ids_points':'count', 'rider_score_points':'mean', "travel_distance_points":'sum'})
best_rider = GroupBy_riders.sort_values(by=['order_ids_points'], ascending=[False]).round(2)

best_rider.loc[(best_rider['order_ids_points'] <= 100), 'order_ids_points'] = best_rider['order_ids_points'] * 3
best_rider.loc[(best_rider['order_ids_points'] > 100) & (best_rider['order_ids_points'] <= 200), 'order_ids_points'] = best_rider['order_ids_points'] * 4
best_rider.loc[(best_rider['order_ids_points'] > 300) & (best_rider['order_ids_points'] <= 400), 'order_ids_points'] = best_rider['order_ids_points'] * 5
best_rider.loc[(best_rider['order_ids_points'] > 400) & (best_rider['order_ids_points'] <= 500), 'order_ids_points'] = best_rider['order_ids_points'] * 6

best_rider.loc[(best_rider['rider_score_points'] <= 4.00), 'rider_score_points'] = best_rider['rider_score_points'] * 100.00
best_rider.loc[(best_rider['rider_score_points'] > 4.00) & (best_rider['rider_score_points'] <= 4.25), 'rider_score_points'] = best_rider['rider_score_points'] * 150.00
best_rider.loc[(best_rider['rider_score_points'] > 4.25) & (best_rider['rider_score_points'] <= 4.50), 'rider_score_points'] = best_rider['rider_score_points'] * 200.00
best_rider.loc[(best_rider['rider_score_points'] > 4.50) & (best_rider['rider_score_points'] <= 4.75), 'rider_score_points'] = best_rider['rider_score_points'] * 300.00
best_rider.loc[(best_rider['rider_score_points'] > 4.75) & (best_rider['rider_score_points'] <= 4.90), 'rider_score_points'] = best_rider['rider_score_points'] * 400.00
best_rider.loc[(best_rider['rider_score_points'] > 4.90) & (best_rider['rider_score_points'] <= 5.0), 'rider_score_points'] = best_rider['rider_score_points'] * 500.00

best_rider['total_points'] = (best_rider['order_ids_points'] + best_rider['rider_score_points'] + best_rider['travel_distance_points'])

best_rider.sort_values(by=['total_points'], ascending=[False]).head(10).round(2)
```
