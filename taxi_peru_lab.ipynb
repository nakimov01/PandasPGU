```python
import pandas as pd
df = pd.read_csv("taxi_peru.csv", delimiter=';', encoding = "Windows-1251")
df
```

```python
df.dtypes
```

```python
df['driver_score'] = df['driver_score']. fillna (0)
df['rider_score'] = df['rider_score']. fillna (0)
df
```

```python
df = df[(df.end_state != 'not found') & (df.end_state != 'rider cancel') & (df.end_state != 'driver cancel')]
df
```

```python
df.info()
```

```python
Sort = df.query("`rider_score` >= 0.0")
a = Sort.groupby(['driver_id']).agg({'journey_id':'count', 'rider_score':'mean'})
a.sort_values(by=['journey_id', 'rider_score'], ascending=[False, False]).round(2)
```

```python

a = df.groupby(['user_id']).agg({'taxi_id':'count', 'driver_id':'count'})
a.sort_values(by=['taxi_id', 'driver_id'], ascending=[False, False]).head(5)
```
