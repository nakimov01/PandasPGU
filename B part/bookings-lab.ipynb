```python
df.query("`Country` == 'GBR' and `Arrival Date Year` == 2016 and `Meal` != 'Undefined'")['Meal'].value_counts().idxmax()
```

```python
Sort = df[(df['Hotel'] == "Resort Hotel") & (df["Country"] == "PRT")]
Sort.groupby(['Arrival Date Month', 'Arrival Date Year']).agg({'Is Canceled':'sum'}).sort_values(['Is Canceled'], ascending=[False]).head(1)
```

```python
Sort = df.query("`Hotel` == 'Resort Hotel' and `Is Canceled` == 1 and `Arrival Date Year` == 2016 ")
Sort.groupby(['Arrival Date Month', 'Arrival Date Year']).agg({'Is Canceled':['sum']})
```

```python
Sort = df.query("`Hotel` == 'Resort Hotel' and `Arrival Date Year` == 2015")
a = Sort.groupby(['Arrival Date Year','Arrival Date Month']).agg({'Is Canceled':'sum',})
a.sort_values(by=[ 'Is Canceled'], ascending=[False]).head(1)
```

```python
Sort = df.query("`Hotel` == 'Resort Hotel' and `Arrival Date Year` == 2016")
a = Sort.groupby(['Arrival Date Year','Arrival Date Month']).agg({'Is Canceled':'sum',})
a.sort_values(by=[ 'Is Canceled'], ascending=[False]).head(1)
```

```python
Sort = df.query("`Hotel` == 'Resort Hotel' and `Arrival Date Year` == 2017")
a = Sort.groupby(['Arrival Date Year','Arrival Date Month']).agg({'Is Canceled':'sum',})
a.sort_values(by=[ 'Is Canceled'], ascending=[False]).head(1)
```

```python
Sort = df.query("`Children` == 0.0 and `Arrival Date Year` == 2015 and `Is Canceled` != 1")
a = Sort.groupby(['Country']).agg({'Children':'count',})
a.sort_values(by=['Children'], ascending=[False]).head(5)
```

```python
df['Total Kids'] = df['Children'] + df['Babies']
df['Total Kids']
```

```python
Sort = df.query("`Total Kids` != 0.0")
a = Sort.groupby(['Country']).agg({'Total Kids':'mean',})
a.sort_values(by=['Total Kids'], ascending=[False]).head(5)
```
